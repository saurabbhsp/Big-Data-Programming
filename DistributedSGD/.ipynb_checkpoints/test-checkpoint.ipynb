```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Data/KDDcup98/Train/cup98LRN.txt")
data.head()
```

```python
featureList = ['STATE', 'PVASTATE', 'MDMAUD', 'CLUSTER', 'GENDER', 'HIT',
              'DATASRCE', 'MALEMILI', 'MALEMILI', 'VIETVETS', 'WWIIVETS', 'LOCALGOV', 'STATEGOV', 'FEDGOV',
              'CARDPROM', 'NUMPROM', 'RAMNTALL', 'NGIFTALL', 'CARDGIFT', 'AVGGIFT']
features = data[featureList]
target = data[['TARGET_D']]
features = pd.get_dummies(features)
features.head()
```

```python
targetClassify = []
for targetValue in target.TARGET_D:
    if targetValue > 0:
        targetClassify.append(1)
    else:
        targetClassify.append(0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, targetClassify, test_size=0.33, random_state=42)
```

```python
def logisticFunction(x):
    return 1.0/(1.0 + np.exp(-x))


def  stochasticGradientAscent(xTrain, yTrain, beta, alpha, 
                                 epochs = 4, epsilon = 1.0e-12):
    """Add bias to x"""
    xTrain = np.insert(xTrain, 0, 1, axis = 1)

    
    """Changing the data type to float"""
    xTrain = xTrain * 1.0
    yTrain = yTrain * 1.0
           
    """Index"""
    indices = np.array(range(0, len(xTrain)))

    for i in range (0, epochs):
        
        """Shuffle the indices"""                        
        np.random.shuffle(indices)
        
        for index in indices:    
            x = xTrain[index]
            y = yTrain[index]
        
            """Calculate probability"""
            betaX = np.dot(beta, x.T)
            p = logisticFunction(betaX)
            
            """Here we are adding gradient as the we are trying to
            maximize the log likelihood and not minimize it"""
            gradient = np.dot(x.T, y - p)
            beta = beta + (alpha * gradient)
        logLikelihood = (yTrain * np.dot(beta, xTrain.T) - np.log(1 + np.exp(np.dot(beta, xTrain.T)))).sum()
        print(logLikelihood)
    return beta
```
